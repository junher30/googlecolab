```python
from google.colab import files
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Subir archivo manualmente desde la computadora
uploaded = files.upload()

# Obtener el nombre del archivo
file_name = list(uploaded.keys())[0]

# Cargar el archivo en un DataFrame
df = pd.read_excel(file_name, sheet_name="Servicios_Prestados")

# Eliminar espacios en los nombres de las columnas para evitar errores
df.columns = df.columns.str.strip()

# Ver las primeras 5 filas
df.head()
df.info()
df.describe()
df.head(100)

# Calcular el gasto total por cliente
gasto_por_cliente = df.groupby("ID_Cliente")["Valor_Pagado"].sum()

# Calcular el gasto promedio por cliente
gasto_promedio = gasto_por_cliente.mean()

# Mostrar resultados
print("Gasto total por cliente:")
print(gasto_por_cliente.head())
print("\nGasto promedio por cliente:", gasto_promedio)

# Visualización del gasto por cliente
plt.figure(figsize=(10, 5))
sns.histplot(gasto_por_cliente, bins=20, kde=True)
plt.title('Distribución del Gasto por Cliente')
plt.xlabel('Gasto Total ($)')
plt.ylabel('Frecuencia')
plt.savefig("histograma_gasto_cliente.png")
plt.show()

# Guardar y descargar archivos
output_csv = "datos_procesados.csv"
df.to_csv(output_csv, index=False)
files.download(output_csv)
files.download("histograma_gasto_cliente.png")

```
